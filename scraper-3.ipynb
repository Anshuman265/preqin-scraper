```python
import os
import json
import pandas as pd
d = {'name':[],'company':[],"Linkedin Url": [],
    "vcard": [],
    "role": [],
    "position": [],
    "location": [],
    "email": [],
    "telephone":[]}
file_dir = 'C:/Internships/morphosis/preqin-scraper/data'
for file in os.listdir(file_dir):
    company = file.split('.')[0]
    with open(os.path.join(file_dir,file), 'r') as f:
        data = json.load(f)
    for person in data:
        d["name"].append(person['name'])
        d["company"].append(company)
        try:
            d["Linkedin Url"].append(person['linkedIn_url'])
        except:
            d["Linkedin Url"].append('-')
        d['vcard'].append(person['vcard'])
        d['role'].append(person['role'])
        d["position"].append(person['position'])
        d["location"].append(person['location'])
        d["email"].append(person['email'])
        d["telephone"].append(person['telephone'])


print(d)

```

```python
df = pd.DataFrame(data=d)
df.head()
```

```python
writer = pd.ExcelWriter('data.xlsx')
df.to_excel(writer)
writer.save()
```
